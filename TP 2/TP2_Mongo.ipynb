```python
# ligne de commande cmd

# mongorestore --db gym C:\Users\David\Desktop\Workspace\Dev\mongodb\Gymnase-mongodb\Gymnases.bson

# mongorestore --db gym C:\Users\David\Desktop\Workspace\Dev\mongodb\Gymnase-mongodb\Sportifs.bson

# 1) Afficher toutes les collections de la base, Compter le nombre de documents ds chacune des collections, Afficher un document de chacune des collections ?

import pymongo
import json
from pprint import pprint

try:
    client = pymongo.MongoClient('mongodb://localhost:27017/')
except:
    print('Authentification failed ')


try :
    db = client.gym
    sportifs = db.Sportifs
    gymnases = db.Gymnases
except :
    print("database introuvable")

# Afficher toutes les collections de la base
print(db.list_collection_names())

# Compter le nombre de documents ds chacune des collections
print(sportifs.count())
print(gymnases.count())

# Afficher un document de chacune des collections
pprint(gymnases.find_one())
pprint(sportifs.find_one())
```

```python
# 2) On cherche le (ou les) conseiller(s) s’appelant "KERVADEC" ?
for sportif in sportifs.find({"Nom" : "KERVADEC"}):
    pprint(sportif)
```

```python
# 3) On cherche les sportifs qui ont ou bien 32 ans ou 40 ans ?
for sportif in sportifs.find({"$or":[{"Age": 32}, {"Age": 40}]}):
    pprint(sportif)
```

```python
# 4) On cherche les sportifs jouant au Basket-ball ?
for sportif in sportifs.find({'Sports.Jouer' : "Basket ball"}):
    pprint(sportif)
```

```python
# 5) On cherche les sportifs ayant au moins 32 ans, ou de sexe féminin ?
for sportif in sportifs.find({"$or":[{"Sexe": "F"}, {"Age": {"$gt":32}}]}):
    pprint(sportif)
```

```python
# 6) Trouver le nombre de sportifs de sexe féminin dans la base ?
for sportif in sportifs.find({'Sexe' : "F"}):
    pprint(sportif)
```

```python
# 7) Quels sont les sportifs (identifiant, nom et prénom) qui ont entre 20 et 30 ans ?
for sportif in sportifs.find({"Age":{"$gte":20,"$lt":30}}):
    pprint(sportif)
```

```python
# 8) Quels sont les sportifs (identifiant et nom) qui pratiquent du hand ball ?
for sportif in sportifs.find({'Sports.Jouer' : "Hand ball"},{"IdSportif" : 1, "Nom" : 1}):
    pprint(sportif)
```

```python
# 9) Quels sont les gymnases de “Villetaneuse” ou de “Sarcelles” qui ont une surface de plus de 400 m2 ?
for gymnase in gymnases.find({"$or":[{"Ville": "VILLETANEUSE"}, {"Ville": "SARCELLES"}], "Surface":{"$gt":400}}):
    pprint(gymnase)
```

```python
# 10) Dans quels gymnases peut-on jouer au hockey le mercredi après 15H ?
for gymnase in gymnases.find({'Seances.0.Libelle' : "Hockey", "Seances.0.Horaire" : {"$gt" : 15.0}}):
    pprint(gymnase)
```

```python
# 11) Quels sportifs (identifiant et nom) ne pratiquent aucun sport ?
for sportif in sportifs.find({"Sports":{"$exists":False}},{"IdSportif" : 1, "Nom": 1}):
    pprint(sportif)
```

```python
# 12) Quels gymnases n’ont pas de séances le dimanche ?
for gymnase in gymnases.find({'Seances.Jour' : { "$ne" : "dimanche"} }):
    pprint(gymnase)
```

```python
# 13) Quels gymnases ne proposent que des séances de basket ball ou de volley ball ?
for gymnase in gymnases.find({"Seances":{"$elemMatch": {"$or":[{"Libelle":'Volley ball'},{"Libelle":'Basket ball'}]}}},{"IdGymnase" : 1, "Seances": 1}):
    pprint(gymnase)
```

```python
# 14) Quels sont les entraîneurs qui sont aussi joueurs ?
for sportif in sportifs.find({"Sports.Jouer":{"$exists":True}, "Sports.Entrainer":{"$exists":True}}):
    pprint(sportif)
```

```python
# 15) Quels sont les sportifs qui sont des conseillers ?
for sportif in sportifs.find({
    'Sports.Jouer' : { "$exists": True },
    'IdSportif' : { '$in' : sportifs.find({ 
        'IdSportifConseiller' : { "$exists": True }
    }).distinct('IdSportifConseiller')}
}):
    pprint(sportifs)
```

```python
# 16)	Pour le sportif “Kervadec” quel est le nom de son conseiller ?
sportif = sportifs.find_one({"Nom" : "KERVADEC"})
pprint(sportifs.find_one({"IdSportif" : sportif['IdSportifConseiller']}))
```

```python
# 17) Quels entraîneurs entraînent du hand ball et du basket ball ?
for sportif in sportifs.find({"$and" : [{"Sports.Entrainer" : "Hand ball"}, {"Sports.Entrainer" : "Basket ball"}]}):    
     pprint(sportif)
```

```python
# 18) Quels sportifs n’ont pas de conseillers ?
for sportif in sportifs.find({"IdSportifConseiller":{"$exists": False}}):    
     pprint(sportif)
```
